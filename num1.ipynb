```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1...Datapoints
x = [1,2,3,4,5]
y = [1,3,2,3,5]
yHat = np.zeros(len(y))
for i in range(len(y)):
    yHat[i] = 0.4 + 0.8 * x[i]

plt.scatter(x,y)
plt.plot(x, yHat, 'g')
plt.grid(True)
plt.show()
```

```python
# 2...Datapoints
x = [1,2,3,4,5]
y = [1,3,2,3,5]
yHat = np.zeros(len(y))
for i in range(len(y)):
    yHat[i] =  x[i]

plt.scatter(x,y)
plt.plot(x, yHat,'r')
plt.grid(True)
plt.show()
```
